```python
import nashpy as nash

import numpy as np

# [[],[],[],[]]
A = np.array([[-1,10,10,10],[0,-1,0,10],[0,10,-1,10],[0,0,0,-1]])
B = np.array([[-1,0,0,0],[0,-1,10,0],[10,0,-1,0],[10,10,10,-1]])

rps = nash.Game(A,B)
rps
```

```python
eqs = rps.support_enumeration()
list(eqs)
```

```python
import numpy as np

iterations = 100

np.random.seed(1)

play_counts = rps.fictitious_play(iterations=iterations)

for row_play_count, column_play_count in play_counts:

    print(row_play_count, column_play_count)
```
